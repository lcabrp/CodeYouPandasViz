### Using matplotlib

```python
# Creating a simple dataset
df = pd.DataFrame({
    'Year': [2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Sales_A': [100, 150, 200, 250, 300, 350, 400], 
    'Sales_B': [90, 130, 180, 220, 280, 350, 310],
})

df.plot(x='Year', y='Sales_A', kind='line')
plt.show()

```

```python
df.plot(x='Year', y=['Sales_A', 'Sales_B'], kind='bar', title="Sales Over Years", grid=True, figsize=(8,5))

plt.show()
```

```python
df.plot(
    x='Year',
    y=['Sales_A', 'Sales_B'],
    kind='line',
    title="Sales Trends",
    grid=True,
    legend=True,
    style=['r--', 'b-'],
    logy=False,
    figsize=(10,6),
    rot=45,
    fontsize=12,
    colormap='viridis')


plt.show()
```

### Using Seaborn

```python
sns.barplot(x='Year', y='Sales_B', data=df, palette='pastel', hue='Year', legend=False)
plt.title('Seaborn Bar Chart')
plt.show()
```

```python
# Line plot
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))
sns.lineplot(x='Year', y='Sales_A', data=df, label='Sales A', marker='o')
sns.lineplot(x='Year', y='Sales_B', data=df, label='Sales B', marker='s')
plt.title("Sales Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.legend()
plt.show()


```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='Year', y='Sales_A', data=df, color="skyblue", label="Sales A")
sns.barplot(x='Year', y='Sales_B', data=df, color="salmon", alpha=0.7, label="Sales B")
plt.title("Yearly Sales Comparison")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
# Notes to myself
# hue='Year' to color the points by year
# size='Year' to make larger points for later years
# palette="coolwarm" to use a color palette that changes with the year

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Sales_A', y='Sales_B', data=df, hue='Year', palette="coolwarm", size='Year', sizes=(50, 200))
plt.title("Sales A vs. Sales B")
plt.xlabel("Sales A")
plt.ylabel("Sales B")
plt.show()
```
