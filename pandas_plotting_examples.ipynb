# **Data Visualization with Pandas, Matplotlib, and Seaborn**

## **1. Introduction**
Data visualization is crucial for understanding trends and patterns. While Pandas provides quick plots, Matplotlib allows fine-grained customization, and Seaborn makes statistical visualizations easy and aesthetically pleasing.

Here's when to use each:
- **Pandas**: Quick exploratory data visualizations using built-in plotting functions.
- **Matplotlib**: Highly customizable, versatile plots with full control over aesthetics.
- **Seaborn**: Ideal for statistical plots with built-in themes and improved readability.

Now, let’s explore examples using these tools.

---

## **2. Setting Up**
Make sure you have the required libraries installed:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

---

## **3. Example Dataset**
Let's create a simple dataset to work with:

```python
# Creating a simple dataset
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [10, 15, 7, 12, 18]
}

df = pd.DataFrame(data)
print(df)
```

---

## **4. Pandas Visualization**
Pandas' built-in plotting uses Matplotlib under the hood. It's ideal for quick exploratory analysis.

```python
df.plot(kind='bar', x='Category', y='Values', title='Pandas Bar Chart')
plt.show()
```

✅ **When to Use Pandas:**  
- For quick plots with minimal customization.
- When working directly with Pandas DataFrames.

🚫 **Limitations:**  
- Limited styling and customization.

---

## **5. Matplotlib Visualization**
Matplotlib offers greater flexibility and customization:

```python
plt.figure(figsize=(6, 4))
plt.bar(df['Category'], df['Values'], color='skyblue')
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Customized Bar Chart with Matplotlib')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

✅ **When to Use Matplotlib:**  
- When you need full control over plot elements.
- For precise styling, annotations, and multi-figure layouts.

🚫 **Limitations:**  
- Requires more code for setup.

---

## **6. Seaborn Visualization**
Seaborn makes statistical data visualization simpler and more aesthetically pleasing:

```python
sns.barplot(x='Category', y='Values', data=df, palette='pastel')
plt.title('Seaborn Bar Chart')
plt.show()
```

✅ **When to Use Seaborn:**  
- For statistical visualizations.
- When working with DataFrames and applying themes.

🚫 **Limitations:**  
- Less control than raw Matplotlib.
- Not always suited for highly customized plots.

---

## **7. Summary Table**
| Library    | Use Case | Strengths | Limitations |
|------------|---------|------------|------------|
| **Pandas** | Quick exploratory plots | Easy, built-in | Limited customization |
| **Matplotlib** | Highly customizable plots | Full control | More setup required |
| **Seaborn** | Statistical visualizations | Beautiful default styles | Less flexible than Matplotlib |

---

## **8. Conclusion**
- Use **Pandas** for quick and simple plots.
- Use **Matplotlib** for highly customized, precise control over elements.
- Use **Seaborn** for stylish and statistical data visualizations.

````python
import pandas as pd
import matplotlib.pyplot as plt

# url to the data set and reading in the data 
url = 'https://raw.githubusercontent.com/CodeYouOrg/DataOpenClass/refs/heads/main/SalaryData.csv'
df = pd.read_csv(url)

# dropping columns we dont need
columns_to_drop = ['Overtime_Rate',	'Incentive_Allowance', 'Other', 'ObjectId']
df = df.drop(columns=columns_to_drop)

# We dont have a ton of missing data so we are dropping the NaNs 
df = df.dropna()

# filtering the data to year 2024
filtered_data = df[df['CalYear'] == 2024]

# Filter out rows where jobTitle contains 'chief' or 'director'... (case-insensitive)
filtered_data = filtered_data[~filtered_data['jobTitle'].str.contains('chief|director|Major|Sergeant|General|Manager|Mayor|Lieutenant|Senior|Commander|Supervisor', case=False, na=False)]

# grouping department together and sorting them 
department_total = filtered_data.groupby(['Department', 'jobTitle'])['YTD_Total'].mean()
department_total = department_total.sort_values(ascending=False)

# Disable scientific notation for pandas
pd.set_option('display.float_format', '{:,.0f}'.format)
print("The top 10 departments by salary spend is:\n", department_total.head(10))


department_total.head(10).plot(kind = 'bar',
                               xlabel = "Department and Title.",
                               ylabel = 'Total Salary',
                               title = 'Total Salary by Department and Title.')
plt.show()
````
